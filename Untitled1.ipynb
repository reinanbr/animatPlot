```python
from scipy.special import gamma
import matplotlib.pyplot as plt
import numpy as np
```

```python
from animateplot import AnimatePlot as ap
```

```python
s = np.linspace(-5,6,400)
g = gamma(s)

g = np.nan_to_num(g,nan=0,posinf=0,neginf=0)

def plot_gamma(i,plt):
    plt.style.use('classic')
    plt.grid(lw=.5,linestyle='--')
    plt.xlim((s.min(),s.max()))
    plt.ylim((g.min(),g.max()))
    plt.plot(s[:i],g[:i])
    plt.title('Gamma Function\n'+fr'$\Gamma({s[i-1]:.2f}) = {g[i-1]:.2f}$'+' | '+fr'$({(s[i-1]-1):.2f})! = {g[i-1]:.2f}$')
    return plt
```

```python
animat = ap(s,plot_gamma)
animat.render_cache()
animat.render_mp4('gamma.mp4',fps=30)
```
