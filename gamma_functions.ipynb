```python
import matplotlib.pyplot as plt
import numpy as np
from animateplot import AnimatePlot as ap
plt.style.use('seaborn')
```

```python
def harmonic(n:int):
    res = 0
    #print(type(n))
    if type(n)==np.ndarray:
        res_list = []
        for ni in n:
            for i in range(1,int(ni+1)):
                res += 1/i
            res_list.append(res)
            res=0
        return np.array(res_list)
    else:
        for i in range(1,n+1):
            res += 1/i
        return res

def euler_mask(N:int):
    return harmonic(N) - np.log(N)


euler_mask(1000)
```

```python
n = np.linspace(1,200,200)

cresc_euler = euler_mask(n)


euler_gamma = [np.euler_gamma for _ in range(200)]

plt.plot(n,cresc_euler,label=r'$\gamma=\lim_{N\to\infty}{\left(H_{N}-ln(N)\right)}$')
plt.plot(n,euler_gamma,label='constante de Euler-Mascheroni '+r'$\gamma$ = '+f'{np.euler_gamma:.4f}...')
plt.annotate(r'$\gamma=\lim_{N\to\infty}{\left(H_{N}-ln(N)\right)}$', fontweight='bold',xytext=(60,.65),xy=(40,np.euler_gamma),size=15,arrowprops = dict(facecolor ='green',
                                  shrink = 0.001))
plt.xlabel(r'$N$', fontweight='bold')
plt.ylabel(r'$\gamma_N$', fontweight='bold')
plt.title(r'provando que a diferena da srie harmnica de $N$ pelo '+'\n'+'logaritmo natural de $N$, com $N$ -> $\infty$,  igual a constante de Euler-Mascheroni',
         fontweight='bold')
plt.legend()
```

```python


def callplot(plt,e,n):
    
    
    plt.plot(n,e,label=r'$\gamma=\lim_{N\to\infty}{\left(H_{N}-ln(N)\right)}$')
    plt.plot(n,euler_gamma[:len(n)],label='constante de Euler-Mascheroni '+r'$\gamma$ = '+f'{np.euler_gamma:.4f}...')
    
    plt.annotate(r'$\gamma=\lim_{N\to\infty}{\left(H_{N}-ln(N)\right)}$', fontweight='bold',xytext=(60,.3),xy=(40,np.euler_gamma),size=15,arrowprops = dict(facecolor ='green',
                                      shrink = 0.001))
    e_txt = 0
    n_txt = 0
    if len(n)>2:
        e_txt = e[-1]
        n_txt = n[-1]
    plt.annotate(fr'${e_txt:.4f}=H({n_txt})-ln({n_txt})$', fontweight='bold',xytext=(40,.9),xy=(7,.7),size=10,arrowprops = dict(facecolor ='blue',
                                      shrink = 0.001))
    
    plt.xlabel(r'$N$', fontweight='bold')
    plt.ylabel(r'$\gamma_N$', fontweight='bold')
    plt.xlim(0,200)
    plt.ylim(0,1.5)
    
    plt.title(r'provando que a diferena da srie harmnica de $N$ pelo '+'\n'+'logaritmo natural de $N$, com $N$ -> $\infty$,  igual a constante de Euler-Mascheroni',
             fontweight='bold')
    plt.legend()
    
    return plt


animate = ap(euler_mask,n,callback_plot=callplot)
animate.render_cache()
animate.render_mp4('euler.mp4')
```

```python

F = 10
m = 5
w = 2


def x(t):
  return (F*t/(w*m))*np.sin(w*t)


t = np.linspace(0,150,150)
```

```python
def callplot(plt,d,t):
    plt.plot(t,d)
    plt.ylabel(r'$x(t)$')
    plt.xlabel(r'$t$')
    plt.title('plotagem da oscilao do sistema entrando em estado de ressonncia',fontweight='bold')
    plt.annotate(r'$\frac{F_0t}{mw_0}\sin(tw_0)$',xy=(15,50),size=15)
    plt.ylim(-120,150)
    plt.xlim(0,150)
    
    return plt


anime = ap(x,t,callback_plot=callplot)

anime.render_cache()
anime.render_mp4('res.mp4')
```
