```python
from noawclg import get_noaa_data as gnd

data_noaa = gnd(date='23/10/2025')

wind = data_noaa['ugrdmwl'].to_numpy()
```

```python
# ... (import e gerao de timestamps como antes)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from animateplot import AnimatePlot

from IPython.display import display

from datetime import datetime, timedelta



def generate_timestamps(quantity=128, interval_hours=3, start_date=None):
    """
    Generates a list of timestamps with specified interval.
    
    Args:
        quantity (int): Number of timestamps to generate (default: 128)
        interval_hours (int): Interval in hours between each timestamp (default: 3)
        start_date (datetime): Initial date/time (default: now)
    
    Returns:
        list: List of timestamps in datetime format
    """
    if start_date is None:
        start_date = datetime.now()
    
    timestamps = []
    for i in range(quantity):
        timestamp = start_date + timedelta(hours=interval_hours * i)
        timestamps.append(timestamp)
    
    return timestamps



timestamps = generate_timestamps(quantity=128, interval_hours=3, start_date=datetime(2025, 10, 19, 0, 0, 0))

fig = plt.figure(figsize=(14, 10), facecolor='black')
m = Basemap(projection='ortho', resolution='l', lat_0=-9, lon_0=-40)

lats = np.linspace(90, -90, 721)
lons = np.linspace(-180, 180, 1440)
lon2d, lat2d = np.meshgrid(lons, lats)
x, y = m(lon2d, lat2d)

vmin, vmax = wind.min(), wind.max()

# mapa base (desenha s uma vez)
m.drawcoastlines(linewidth=0.5)
m.drawcountries(linewidth=0.3)
m.fillcontinents(color='lightgray', lake_color='white')
m.drawmapboundary(fill_color='white')
m.drawmeridians(np.arange(-180, 181, 60), linewidth=0.5, color='gray', alpha=0.5)
m.drawparallels([0], linewidth=2, color='red', dashes=[1,0])
m.drawparallels([23.5, -23.5], linewidth=.5, color='orange', dashes=[5,3])
m.drawparallels([66.5, -66.5], linewidth=.5, color='blue', dashes=[5,3])
m.drawparallels(np.arange(-90, 91, 30), linewidth=0.1, color='gray', alpha=0.5)

# cria o QuadMesh inicial com pcolormesh
# importante: usar wind[0].ravel() ao atualizar, mas aqui passamos a matriz normal
mesh = m.contourf(x, y, wind[0], shading='auto', cmap='magma', vmin=vmin, vmax=vmax)


cbar = plt.colorbar(mesh, orientation='horizontal', pad=0.03, shrink=0.5, aspect=30)
cbar.set_label('Velocidade do Vento (m/s)', fontsize=9, color='white')
cbar.ax.tick_params(labelsize=8, colors='white')

title = plt.title('', fontsize=14, fontweight='bold', color='white')

def plot_wind(it, mesh):
    # Atualiza os dados do QuadMesh. Para QuadMesh, set_array espera um array 1D com os dados
    mesh = m.contourf(
        x, y, wind[it],
        shading='auto',
        cmap='magma',
        vmin=vmin, vmax=vmax
    )
    # se precisar atualizar limites dinamicamente:
    # mesh.set_clim(vmin, vmax)
    title.set_text(f'Jatos de Vento \n {timestamps[it].strftime("%d %B %Y %H:%M")}')
    return plt

animator = AnimatePlot(space=list(range(len(timestamps))), callplot=plot_wind, dpi=120, clf=False)
animator.render_cache()
animator.render_gif('test.gif')
# resto do fluxo (cache/render) igual
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from animateplot import AnimatePlot
from datetime import datetime, timedelta

class DataAnimator:
    def __init__(self, wind, start_date=datetime(2025, 10, 19, 0, 0, 0), quantity=128, interval_hours=3):
        self.wind = wind
        self.start_date = start_date
        self.quantity = quantity
        self.interval_hours = interval_hours
        
        # Inicializao modular
        self.timestamps = self._setup_timestamps()
        self.fig = self._setup_figure()
        self.map = self._setup_basemap()
        self.lats, self.lons = self._setup_coordinates()
        self.lon2d, self.lat2d = self._setup_meshgrid()
        self.x, self.y = self._setup_projection()
        self.vmin, self.vmax = self._setup_colorscale()
        self._draw_map_features()
        self.mesh = self._setup_initial_contour()
        self.cbar = self._setup_colorbar()
        self.title = self._setup_title()
    
    def _setup_timestamps(self):
        """Gera lista de timestamps baseada na data inicial, quantidade e intervalo."""
        timestamps = []
        for i in range(self.quantity):
            timestamps.append(self.start_date + timedelta(hours=self.interval_hours * i))
        return timestamps
    
    def _setup_figure(self):
        """Configura a figura matplotlib."""
        return plt.figure(figsize=(14, 10), facecolor='black')
    
    def _setup_basemap(self):
        """Configura o mapa base com projeo ortogrfica."""
        return Basemap(projection='ortho', resolution='l', lat_0=-9, lon_0=-40)
    
    def _setup_coordinates(self):
        """Define arrays de latitudes e longitudes."""
        lats = np.linspace(90, -90, 721)
        lons = np.linspace(-180, 180, 1440)
        return lats, lons
    
    def _setup_meshgrid(self):
        """Cria malha 2D de coordenadas."""
        return np.meshgrid(self.lons, self.lats)
    
    def _setup_projection(self):
        """Projeta coordenadas geogrficas para coordenadas do mapa."""
        return self.map(self.lon2d, self.lat2d)
    
    def _setup_colorscale(self):
        """Define limites mnimo e mximo para escala de cores."""
        return self.wind.min(), self.wind.max()
    
    def _draw_map_features(self):
        """Desenha caractersticas geogrficas no mapa."""
        m = self.map
        m.drawcoastlines(linewidth=0.5)
        m.drawcountries(linewidth=0.3)
        m.fillcontinents(color='lightgray', lake_color='white')
        m.drawmapboundary(fill_color='white')
        m.drawmeridians(np.arange(-180, 181, 60), linewidth=0.5, color='gray', alpha=0.5)
        m.drawparallels([0], linewidth=2, color='red', dashes=[1,0])
        m.drawparallels([23.5, -23.5], linewidth=.5, color='orange', dashes=[5,3])
        m.drawparallels([66.5, -66.5], linewidth=.5, color='blue', dashes=[5,3])
        m.drawparallels(np.arange(-90, 91, 30), linewidth=0.1, color='gray', alpha=0.5)
    
    def _setup_initial_contour(self):
        """Cria o contorno inicial com os primeiros dados."""
        return self.map.contourf(
            self.x, self.y, self.wind[0],
            shading='auto', cmap='magma',
            vmin=self.vmin, vmax=self.vmax
        )
    
    def _setup_colorbar(self):
        """Configura a barra de cores."""
        cbar = plt.colorbar(
            self.mesh,
            orientation='horizontal',
            pad=0.03,
            shrink=0.5,
            aspect=30
        )
        cbar.set_label('Wind Speed (m/s)', fontsize=9, color='white')
        cbar.ax.tick_params(labelsize=8, colors='white')
        return cbar
    
    def _setup_title(self):
        """Configura o ttulo inicial do grfico."""
        return plt.title('', fontsize=14, fontweight='bold', color='white')
    
    def _plot_frame(self, i, mesh):
        """Atualiza o frame da animao."""
        self.mesh = self.map.contourf(
            self.x, self.y, self.wind[i],
            shading='auto', cmap='magma',
            vmin=self.vmin, vmax=self.vmax
        )
        self.title.set_text(f'Jet Streams\n{self.timestamps[i].strftime("%d %B %Y %H:%M")}')
        return plt
    
    def render_gif(self, path='wind.gif', dpi=120):
        """Renderiza animao como GIF."""
        animator = AnimatePlot(
            space=list(range(len(self.timestamps))),
            callplot=self._plot_frame,
            dpi=dpi,
            clf=False
        )
        animator.render_cache()
        animator.render_gif(path)
    
    def render_video(self, path='wind.mp4', dpi=120, fps=24):
        """Renderiza animao como vdeo MP4."""
        animator = AnimatePlot(
            space=list(range(len(self.timestamps))),
            callplot=self._plot_frame,
            dpi=dpi,
            clf=False
        )
        animator.render_cache()
        animator.render_video(path, fps=fps)
```
