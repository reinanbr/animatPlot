```python
import matplotlib.pyplot as plt
from scipy.constants import k
import numpy as np
from animateplot import AnimatePlot as ap
plt.style.use(['dark_background'])
from math import sqrt
```

```python
def is_prime(n):
    if n % 2 == 0:
        return False

    for i in range(3, n):
        if n % i == 0:
            return False

    return True



def gen_prime(N=100):
    space = np.arange(1,N)
    prime_space = []
    for x in space:
        if is_prime(x):
            prime_space.append(x)
    return np.array(prime_space)
    
```

```python
gen_prime(200)
```

```python
def gauss_percent_prime(N=200):
    if type(N)==int:
        return N/np.log(N)
    else:
        res = []
        for n in N:
            res.append(n/(np.log(n)))
    return np.array(res)

    
def percent_prime(N=200):
    if type(N)==int:
        return N/len(gen_prime(N))
    else:
        res = []
        for n in N:
            res.append(len(gen_prime(n)/n))
        return np.array(res)
```

```python
n = np.arange(10,1000,10,dtype=int)

p = percent_prime(n)
g = gauss_percent_prime(n)





plt.title('quantidade de nmeros primos',fontweight='bold')
plt.xlabel('linha dos nmeros inteiros',fontweight='bold')
plt.ylabel('quantidade de nmeros primos',fontweight='bold')
plt.plot(n,p,c='g',label='real quantidade de nmeros primos at x')
plt.plot(n,g,c='m',label='relao de Gauss para a quant. de nm. primos at x')
plt.grid(lw=.2)
plt.legend()

plt.annotate(r'$\pi(x)$',
             xy=(540,100),xytext=(400,120),
             size=12,fontweight='bold',
             arrowprops = dict(facecolor ='g',
                               lw=1.2,
                                  shrink = 0.001))

plt.annotate(r'$\frac{x}{\ln(x)}$',
             xy=(650,100),xytext=(800,80),
             size=16,fontweight='bold',
             arrowprops = dict(facecolor ='m',
                               lw=1.2,
                                  shrink = 0.001))
plt.xlim((0,1000))
plt.ylim((1,170))
```

```python
#criando a animao 1

def callplot_1(x,plt):
    i = len(x)
    plt.title('quantidade de nmeros primos',fontweight='bold')
    plt.xlabel('linha dos nmeros inteiros',fontweight='bold')
    plt.ylabel('quantidade de nmeros primos',fontweight='bold')
    plt.plot(n[:i],p[:i],c='g',label='real quantidade de nmeros primos at x')
    plt.plot(n[:i],g[:i],c='m',label='relao de Gauss para a quant. de nm. primos at x')
    plt.grid(lw=.2)
    plt.legend()

    plt.annotate(r'$\pi(x)$',
                 xy=(540,100),xytext=(400,120),
                 size=12,fontweight='bold',
                 arrowprops = dict(facecolor ='g',
                                   lw=1.2,
                                      shrink = 0.001))

    plt.annotate(r'$\frac{x}{\ln(x)}$',
                 xy=(650,100),xytext=(800,80),
                 size=16,fontweight='bold',
                 arrowprops = dict(facecolor ='m',
                                   lw=1.2,
                                      shrink = 0.001))
    plt.xlim((0,1000))
    plt.ylim((1,170))
    
    return plt

animate_1 = ap(n,callplot=callplot_1)
animate_1.render_cache()
```
