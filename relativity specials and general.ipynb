```python
import matplotlib.pyplot as plt
from scipy.constants import k
import numpy as np
from animateplot import AnimatePlot as ap
plt.style.use('dark_background')
```

## specials

### kinetic energy

```python
m = 1
c = 3*1e8
v = np.linspace(0,.94*c,200)


def er(v): 
    beta = v/c
    y_ = 1/np.sqrt(1-beta**2)
    e0 = m*c**2
    return e0*(y_-1)

def ec(v): return m*v**2/2



er_ = er(v)
ec_ = ec(v)

v_c = v/c
plt.title('enegia cin√©tica: Relativ√≠stica x Cl√°ssica',fontweight='bold')
plt.plot(v/c,er_,label='energia cin. relativ.')
plt.plot(v/c,ec_,label='energia cin. cl√°ssica')
plt.xticks(v_c[::40],[f'{i:.1}c' for i in v_c[::40]])
plt.ylabel('energia (j)')
plt.xlabel('%c (m/s)')
plt.ylim(0,er_.max())
plt.xlim(0,v_c.max())
plt.legend()
plt.grid('--',linewidth=.2)
plt.annotate(r'$E_{relat.}=m_{0}c^{2}(\gamma - 1)$',size=15,xy=(0.05,1.25*1e17))
plt.annotate(r'$E_{cl√°ss.}=\frac{m_{0}v^{2}}{2}$',size=15,xy=(0.05,0.9*1e17))
```

```python
def call_plt(x,f,plt):
    loc = len(x)
    plt.title('enegia cin√©tica: Relativ√≠stica x Cl√°ssica',fontweight='bold')
    plt.plot(v_c[:loc],er_[:loc],label='energia cin. relativ.')
    plt.plot(v_c[:loc],ec_[:loc],label='energia cin. cl√°ssica')
    plt.xticks(v_c[::40],[f'{i:.1}c' for i in v_c[::40]])
    plt.ylabel('energia (j)')
    plt.xlabel('velocidade em %c (m/s)')
    plt.ylim(0,er_.max())
    plt.xlim(0,v_c.max())
    plt.legend()
    plt.grid('--',linewidth=.2)
    plt.annotate(r'$E_{relat.}=m_{0}c^{2}(\gamma - 1)$',size=15,xy=(0.05,1.25*1e17))
    plt.annotate(r'$E_{cl√°ss.}=\frac{m_{0}v^{2}}{2}$',size=15,xy=(0.05,0.9*1e17))
    
    return plt

```

```python
animate = ap(v_c,er,callplot=call_plt)
animate.render_cache()

animate.render_mp4('kinetic_rel.mp4')
```
