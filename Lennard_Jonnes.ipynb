```python
import matplotlib.pyplot as plt
import numpy as np
from animateplot import AnimatePlot as ap
from scipy import special
import matplotlib
plt.style.use('classic')
plt.style.use('dark_background')
```

```python
def strong_force(r,theta): return (theta/r)**12

def small_force(r,theta): return -(theta/r)**6

def lenner_jhonnes(r,theta=.7,e=.5): return e*(strong_force(r,theta) + small_force(r,theta))
```

```python
r = np.linspace(.03,.1,200)
theta=.042
e = .9
v = lenner_jhonnes(r,theta,e)

sv = strong_force(r,theta)
smv = small_force(r,theta)
```
